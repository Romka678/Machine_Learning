# Нормализация признаков

```python
X0 = data_raw.drop('Genre', axis = 1)
y = data_raw['Genre']
print("x:", X0.shape)
print("y:", y.shape)
```

```python
data_raw.info()
```

```python
X1 = X0.drop('Name', axis = 1)
```

```python
X2 = X1.drop('Platform', axis = 1)
```

```python
X3 = X2.drop('Publisher', axis = 1)
```

```python
X4 = X3.drop('Rating', axis = 1)
```

```python
X = X4.drop('Developer', axis = 1)
```

Убрали данные столбцы,так как они не влияют на выборку,их не будем учитывать.

```python
X
```

```python
X = (X - X.mean(axis = 0))/X.std(axis = 0)
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 5)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
trees_array = np.array([10,20,40,80,160])
for i in range (5):
    rfc = RandomForestClassifier(n_estimators = trees_array[i])
    rfc.fit(X_train, y_train)
    print('trees =',trees_array[i],'train accuracy =',rfc.score(X_train, y_train),'test accuracy =', rfc.score(X_test, y_test))
    i+=1
```

Метод RandomForest показывает самую высокую точность на обучаещей выборке для 160 деревьев - 86%.На тестовой самая высокая точность достигается при 160 деревьев - 29%.

```python
from sklearn.neighbors import KNeighborsClassifier
neighbors_array = np.array([1,10,20,30,40,50])
for i in range (6):
    knn = KNeighborsClassifier(n_neighbors = neighbors_array[i])
    knn.fit(X_train, y_train)
    print('neighbours =',neighbors_array[i],'train accuracy =',knn.score(X_train, y_train),'test accuracy =', knn.score(X_test, y_test))
    i+=1
```

kNN метод показывает самую высокую точность на обучаещей выборке для 1 соседа - 84%. При увеличении количества соседей, точность на обучающей выборке падает, а на тестовой самая высокая точность достигается при 50 соседях - 28%,увеличивается с увеличением количества соседей.

# Выводы:

Метод kNN показал меньшую точность,чем Random_Forest.

# Метод главных компонент (PCA)

Применим метод для сокращения размерности пространства признаков,сократим размерность пространства до 2.

```python
from sklearn.decomposition import PCA

sc = sklearn.preprocessing.StandardScaler()
X_standart = sc.fit_transform(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_standart)
X_pca = pd.DataFrame(X_pca, columns=['1 компонента','2 компонента'])
```

```python
X.shape, X_pca.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.3, random_state = 5)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
def biplot(XX, y, coeff, labels):
    plt.figure(figsize=(16,14))
    draw = pd.concat([XX, y], axis=1, join='inner')
    np.random.seed(10)
    random_subset = np.random.choice(np.arange(draw.shape[0]), size=5000, replace=False)
    sns.scatterplot(data=draw.iloc[random_subset], x='1 компонента', y='2 компонента', hue='Genre', alpha=0.7)
    for i in range(0, len(coeff[0])):
        plt.text(coeff[0, i]*5, coeff[1, i]*5, labels[i], color='black', ha='center', va='center')
        plt.plot([0, coeff[0, i]*10], [0, coeff[1, i]*10])
    plt.ylim((-6,7.5))
    plt.xlim((-5,7.5))
    
    
biplot(X_pca, y, pca.components_, X.columns)
```

Как видно, наиболее важными признаками являются:
1)Оценки критиков
2)Продажи во всем мире

# Random Forest

Применим классификатор Random Forest, так как он показывал более высокие результаты до применения PCA.

```python
from sklearn.ensemble import RandomForestClassifier
trees_array = np.array([10,20,40,80,160])
for i in range (5):
    rfc = RandomForestClassifier(n_estimators = trees_array[i])
    rfc.fit(X_train, y_train)
    print('trees =',trees_array[i],'train accuracy =',rfc.score(X_train, y_train),'test accuracy =', rfc.score(X_test, y_test))
```

Точность на тренировочной и тестововй выборках сильно отличаются, это указывает на переобучение.
Добавим параметр max_depth и посмотрим как изменятся результаты.

```python
from sklearn.ensemble import RandomForestClassifier
trees_array = np.array([10,20,40,80,160])
for i in range (5):
    rfc = RandomForestClassifier(n_estimators = trees_array[i],max_depth=5)
    rfc.fit(X_train, y_train)
    print('trees =',trees_array[i],'train accuracy =',rfc.score(X_train, y_train),'test accuracy =', rfc.score(X_test, y_test))
```

Разница точности на тестовой и обучающей выборке стала меньше за счет ограничения максимальной глубины дерева.

Применение PCA понизило точность предсказания, при использовнии двух компонент с 29% до 22.8%

# PCA с различным числом компонент

Построим графики, показывающие ошибку на тестовой и обучающей выборке при использовании разного количества компонент:

```python
from sklearn.ensemble import RandomForestClassifier
test_error = []
train_error = []

pca_array = np.array([j+2 for j in range(9)])

for i in range(9):
    pca = PCA(n_components = pca_array[i])
    X_pca = pca.fit_transform(X_standart)
    X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.25, random_state=10)
    
    rfc = RandomForestClassifier(n_estimators = 200, max_depth=10)
    rfc.fit(X_train, y_train)
    
    test_error.append(1 - rfc.score(X_test, y_test))
    train_error.append(1 - rfc.score(X_train, y_train))

plt.plot(range(2,11), train_error, label="train error", color='red')
plt.plot(range(2,11), test_error, label="test error", color='blue')
plt.xlabel("Components")
plt.ylabel("Error")
plt.legend()
pass
```

Видно,что ошибка на тестовой и тренировочной выборках, при увеличении числа используемых компонент,уменьшается.
